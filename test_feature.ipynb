```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### 测试数据插入特征

```python
train = pd.read_csv('trainSaleDate.csv')
test = pd.read_csv('yancheng_testA_20171225.csv')
trainFeature = pd.read_csv('train_feature.csv')
```

```python
test['year'] = test['predict_date'].apply(lambda x: int(str(x)[0:4]))
test['mouth'] = test['predict_date'].apply(lambda x: int(str(x)[4:6]))
test = test.drop(['predict_date'], axis=1)
test
```

```python
sellMouth = train[['class_id', 'how_many_mouth_has_sell', 'sale_quantity']][(train['year']==2017)&(train['mouth']==10)]
sellMouth.dropna(inplace=True)
sellMouth['how_many_mouth_has_sell'] = sellMouth['how_many_mouth_has_sell']+1
sale_quantity_total_last_mouth = sellMouth['sale_quantity'].sum()
sellMouth.rename(columns={'sale_quantity': 'sale_quantity_this_last_mouth'}, inplace=True)

sellYear = train[['class_id', 'sale_quantity']][((train['year']==2016)&(train['mouth']>=11))|((train['year']==2017)&(train['mouth']<11))]
sellYear.dropna(inplace=True)
sellYear = sellYear.groupby('class_id').agg('sum').reset_index()
sale_quantity_total_last_year = sellYear['sale_quantity'].sum()
sellYear.rename(columns={'sale_quantity': 'sale_quantity_this_last_year'}, inplace=True)

test = pd.merge(test, trainFeature, on='class_id', how='left')
test = pd.merge(test, sellMouth, on='class_id', how='left')
test = pd.merge(test, sellYear, on='class_id', how='left')

test.loc[:, 'sale_quantity_total_last_mouth'] = sale_quantity_total_last_mouth
test.loc[:, 'sale_quantity_total_last_year'] = sale_quantity_total_last_year
test.loc[:, 'this_mouth_sale_class_count'] = 140

test.sort(['class_id'])
test.fillna(0.0, inplace=True)

test = test.drop(['predict_quantity'], axis=1)
# test.to_csv('test_feature.csv', index=False)
test
```
